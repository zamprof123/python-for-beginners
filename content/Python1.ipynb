# First Data Structure: the List

A "Data structure" in computer science is some way of storing information. The simplest one is a list.
In Python you write it very simply, using square brackets, with commas to separate the elements.

```
my_list = [1,2,3,5]
```

If I want to get the 3rd element of the list, I simply write

```
my_list[3]
```

...except not quite. In Python, you start counting from 0 when indexing in a list
(whereas in R you count from 1). So 

```
print(my_list[0]) 
```

would print a 1, and 

```
print(my_list[3])
```

would print a 5. 

What can lists contain in Python? Numbers, strings (ie text), or even other lists.

```
my_list=[1,"zam",[2,3]]
```

is a legitimate list in Python.

## How do I print everything in a list? the "for loop"

The simplest idiom for this is as follows. Let's say we have a list called Y
```
 for item in Y:
 
    print(item)
```
Python sees the "for item in Y" and will go through that list Y, one element at a time, and
each time will set a "local" or "temporary" variable item= that element, and then do whatever comes next. 
This is easier if you try: **Just press Play on the following, and try to compare the code with what comes out and confirm that you understand why.**


````python
print("this is a simple example:")
my_list = [1, 2, 3]
for i in my_list:
    print(i)

# Let's make a little space
print("\n\n")

print("reminder that elements of a list do not all need to be the same type of thing:")
my_list = [1, "superman", 34.5]
for j in my_list:
    print(j)

# Let's make a little space
print("\n\n")
print("also, this is what happens if one of the elements of the list, is itself a list.")
print("I'm just showing you this for completeness, I don't really see why you would need it")
my_list = [1, 2, [100, 37]]
for bigdog in my_list:
    print(bigdog)
````

# How do I try out a while loop, like we saw on the slides just before

Here are 2 examples of how we can use a while loop. 

Again, this is a **very very common idiom** - repeatedly looping over
the same code, and each time you increment a counter.

````python
# While-loop Example 1

print("Simplest example, just keep going til the counter hits a limit\n")
count = 1
while count <= 5:  # This is going to keep looping as long as count is 5 or less
    print(count)
    count = count + 1  # Increment count by 1

# While-loop Example 2

print("show how to escape a loop if a condition is met\n")
number = 77
while True:  # Infinite loop, True is always True
    print(number)
    if number == 73:  # Stop the loop when number reaches 73
        break
    number = number - 1
````
